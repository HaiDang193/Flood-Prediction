```python
import pandas as pd
```

```python
label = pd.read_csv("../data/train/label_raw.csv")

label.head
```

```python
label = label[
    label["Location"].str.contains(
        "Thua Thien Hue|Quang Nam",
        case = False,
        na=False
    )
].copy()
label.head()
```

```python
label["Start Day"] = (
    label["Start Day"]
    .fillna(1)
    .astype(int)
)

label["End Day"] = (
    label["End Day"]
    .fillna(1)
    .astype(int)
)
```

```python
label["start_date"] = pd.to_datetime(
    dict(
        year=label["Start Year"],
        month=label["Start Month"],
        day=label["Start Day"]
    )
)
```

```python
label["end_date"] = pd.to_datetime(
    dict(
        year=label["End Year"],
        month=label["End Month"],
        day=label["End Day"]
    )
)
```

```python
label = label[
    [
        "DisNo.",
        "Location",
        "Latitude",
        "Longitude",
        "start_date",
        "end_date"
    ]
]

label.head()
```

```python
rows = []

for _, row in label.iterrows():
    provinces = str(row["Location"]).split(",")
    for province in provinces:
        rows.append({
            "province": province.strip(),
            "start_date": row["start_date"],
            "end_date": row["end_date"]
        })

label = pd.DataFrame(rows)

label.head()
```

```python
label = label[
    label["province"].isin(
        [
            "Thua Thien Hue",
            "Quang Nam"
        ]
    )
]

label
```

```python
rows = []
for _, row in label.iterrows():
    dates = pd.date_range(
        row["start_date"],
        row["end_date"]
    )
    for d in dates:
        rows.append({
            "date": d,
            "province": row["province"],
            "flood": 1
        })

label_daily = pd.DataFrame(rows)

label_daily.head()
```

```python
label_daily = label_daily.drop_duplicates(
    subset=[
        "date",
        "province"
    ]
)

label_daily = label_daily.sort_values(
    [
        "province",
        "date"
    ]
)

label_daily.head()
```

```python
df = pd.read_csv("../data/train/master_weather.csv")
df.head()
```

```python
df["date"] =pd.to_datetime(df["date"])
```

```python
label_daily.rename(columns={"province":"location"},inplace = True)

label_daily.replace({
    "Thua Thien Hue":"Hue"
},inplace = True)

label_daily.head()
```

```python
master_data = df.merge(
    label_daily[
        [
            "date",
            "location",
            "flood",
        ]
    ],
    on=["date","location"],
    how="left"
)

master_data.head()
```

```python
master_data["flood"] = master_data["flood"].fillna(0)

master_data.head()
```

```python
master_data.describe()
```

```python
master_data.to_csv("../data/train/data_train.csv", index=False)
```
