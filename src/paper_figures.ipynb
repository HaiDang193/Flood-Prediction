```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/train/data_train.csv")

plt.figure(figsize=(6,4))
df["flood"].value_counts().sort_index().plot(kind="bar")

plt.xticks([0,1], ["Non-flood", "Flood"], rotation=0)
plt.xlabel("Label")
plt.ylabel("Count")
plt.title("Flood and Non-Flood Label Distribution")
plt.tight_layout()
plt.show()
```

```python
df["date"] = pd.to_datetime(df["date"])

monthly = (
    df.groupby(["location", "month"])["rainfall"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(8,5))

for loc in monthly["location"].unique():
    temp = monthly[monthly["location"] == loc]
    plt.plot(temp["month"], temp["rainfall"], marker="o", label=loc)

plt.xlabel("Month")
plt.ylabel("Average Rainfall")
plt.title("Monthly Rainfall Pattern by Location")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

df.boxplot(
    column="river_discharge",
    by="flood",
    grid=False
)

plt.suptitle("")
plt.title("River Discharge under Flood and Non-Flood Conditions")
plt.xlabel("Flood Label")
plt.ylabel("River Discharge")
plt.xticks([1,2], ["Non-flood", "Flood"])
plt.tight_layout()
plt.show()
```
