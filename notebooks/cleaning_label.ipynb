```python
import pandas as pd
import numpy as np
```

```python
label = pd.read_csv("../data/label_raw/label_raw.csv")

label.head()
```

```python
label.columns
```

```python
label = label[
    [
        "Location",
        'Start Year',
        'Start Month',
        'Start Day',
        'End Year', 
        'End Month',
        'End Day',
    ]
]

label.head()
```

```python
label.info()
```

```python
label.describe()
```

```python
label = label[
    label["Location"].str.contains(
        "Thua Thien Hue|Quang Nam",
        case = False,
        na=False
    )
].copy()
label.head()
```

```python
label["Start Day"] = (
    label["Start Day"]
    .fillna(1)
    .astype(int)
)

label["End Day"] = (
    label["End Day"]
    .fillna(1)
    .astype(int)
)
```

```python
label["start_date"] = pd.to_datetime(
    dict(
        year=label["Start Year"],
        month=label["Start Month"],
        day=label["Start Day"]
    )
)
```

```python
label["end_date"] = pd.to_datetime(
    dict(
        year=label["End Year"],
        month=label["End Month"],
        day=label["End Day"]
    )
)
```

```python
label.head()
```

```python
rows = []

for _, row in label.iterrows():
    locations = str(row["Location"]).split(",")
    for location in locations:
        rows.append({
            "location": location.strip(),
            "start_date": row["start_date"],
            "end_date": row["end_date"]
        })

label = pd.DataFrame(rows)

label.head()
```

```python
label = label[
    label["location"].isin(
        [
            "Thua Thien Hue",
            "Quang Nam"
        ]
    )
]

label
```

```python
rows = []
for _, row in label.iterrows():
    dates = pd.date_range(
        row["start_date"],
        row["end_date"]
    )
    for d in dates:
        rows.append({
            "date": d,
            "location": row["location"],
            "flood": 1
        })

label_daily = pd.DataFrame(rows)

label_daily.head()
```

```python
label_daily["location"] = label_daily["location"].replace(
    "Thua Thien Hue",
    "Hue"
)
```

```python
label_daily.duplicated().sum()
```

```python
label_daily = label_daily.drop_duplicates(
    subset=[
        "date",
        "location"
    ]
)

label_daily = label_daily.sort_values(
    [
        "location",
        "date"
    ]
)

label_daily.head()
```

```python
label_daily["location_id"] = label_daily["location"].map({
    "Hue":1,
    "Quang Nam":2
}).astype("Int64")

label_daily.head()
```

```python
label_daily = label_daily.drop(columns = ['location'])
label_daily.head()
```

```python
label_daily.to_csv("../data/cleaned/label_cleaned.csv",index=False)
```
